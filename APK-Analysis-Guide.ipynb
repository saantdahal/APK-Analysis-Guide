# Android & Flutter APK Deep Analysis Guide

**Scope:** A practical, phase-by-phase workflow for analyzing your own Android/Flutter APKs useful for debugging, migration planning, architecture review, and security auditing.

> ⚠️ **Ethical use only:** Only analyze applications you own or have explicit permission to inspect.


## Quick Reference

| Phase | Focus | Primary Tool(s) |
|---|---|---|
| 1 | Decode the APK | APKTool |
| 2 | Understand folder structure | — |
| 3 | Java/Kotlin code review | JADX GUI |
| 4 | Flutter asset inspection | File explorer |
| 5 | Native library analysis | Ghidra / IDA Free |
| 6 | Manifest security review | Manual / MobSF |
| 7 | Network traffic analysis | HTTP Toolkit / Burp Suite / mitmproxy |
| 8 | Runtime/log analysis | ADB |
| 9 | Local storage inspection | ADB / File explorer |
| 10 | Database review | DB Browser for SQLite |
| 11 | Final security checklist | — |


## Phase 1 — Decode the APK

**Tool:** APKTool

Run the command below (edit the cell and execute it in an environment where APKTool is installed):

````python
%%bash
apktool d app.apk -o output
````

**Resulting structure:**
```
output/
├── AndroidManifest.xml
├── res/
├── assets/
├── lib/
├── smali/
└── apktool.yml
```


## Phase 2 — Folder Overview

| Folder / File | Purpose |
|---|---|
| `AndroidManifest.xml` | Package name, permissions, activities, services |
| `res/` | Icons, drawables, XML resources, strings |
| `assets/` | Flutter assets, fonts, JSON, images |
| `lib/` | Native libraries (`libapp.so`, `libflutter.so`) |
| `smali/` | Decompiled Dalvik bytecode |
| `apktool.yml` | APKTool configuration metadata |


## Phase 3 — Java/Kotlin Analysis

**Tool:** JADX GUI

**Search terms to check for:**
- `https://` / `http://`
- `api`
- `token`
- `secret`
- `password`
- `Authorization`
- `Firebase`
- `OneSignal`
- `Retrofit`
- `OkHttp`

**Components to review:**
- Activities
- Services
- Broadcast Receivers
- WorkManager jobs
- Firebase initialization code


## Phase 4 — Flutter Analysis

**Inspect:** `assets/flutter_assets/`

**Look for:**
- JSON configuration files
- Fonts
- Images
- Localization files
- App-specific config files


## Phase 5 — Native Library Analysis

**Tools:** Ghidra, IDA Free

**Target files:**
- `libapp.so`
- `libflutter.so`

**Check for:**
- Embedded strings
- Exported symbols
- Native method implementations
- Cryptography usage


## Phase 6 — Manifest Security Review

**Manifest attributes to check:**
- `android:exported`
- `allowBackup`
- `debuggable`
- `usesCleartextTraffic`
- `networkSecurityConfig`

**Sensitive permissions to review:**
- `INTERNET`
- `CAMERA`
- `LOCATION`
- `RECORD_AUDIO`
- `STORAGE`


## Phase 7 — Network Analysis

**Tools:** HTTP Toolkit, Burp Suite Community, mitmproxy

**Review:**
- API requests and endpoints
- JWTs / auth tokens
- Request/response headers
- SSL/TLS configuration
- Certificate pinning implementation


## Phase 8 — Runtime Analysis

**Tool:** ADB

Run the command below in a terminal with a connected device/emulator:

````python
%%bash
adb logcat
````

**Useful log filters:**
- `Flutter`
- `Firebase`
- `OkHttp`
- `OneSignal`


## Phase 9 — Local Storage Inspection

**Inspect:**
- SharedPreferences
- SQLite databases
- Cache directories
- General app files

**Verify these aren't stored unencrypted:**
- Tokens
- Passwords
- Other sensitive information


## Phase 10 — Database Review

**Tool:** DB Browser for SQLite

**Review:**
- Cached data
- User information
- Stored tokens


## Phase 11 — Final Security Checklist

- [ ] Hardcoded API keys
- [ ] Hardcoded passwords
- [ ] Plain-text tokens
- [ ] Debuggable build left enabled
- [ ] Unnecessarily exported components
- [ ] SSL/certificate pinning present
- [ ] Backup enabled (`allowBackup`)
- [ ] Cleartext HTTP traffic allowed
- [ ] WebView debugging enabled
- [ ] Sensitive data in logs
- [ ] Local storage encrypted
