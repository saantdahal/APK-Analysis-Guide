<a id="phase-1"></a>
## Phase 1 — Decode the APK

**Tool:** APKTool

Run the command below (edit the cell and execute it in an environment where APKTool is installed):

````python
%%bash
apktool d app.apk -o output
````

**Resulting structure:**
```
output/
├── AndroidManifest.xml
├── res/
├── assets/
├── lib/
├── smali/
└── apktool.yml
```

<a id="phase-2"></a>
## Phase 2 — Folder Overview

| Folder / File | Purpose |
|---|---|
| `AndroidManifest.xml` | Package name, permissions, activities, services |
| `res/` | Icons, drawables, XML resources, strings |
| `assets/` | Flutter assets, fonts, JSON, images |
| `lib/` | Native libraries (`libapp.so`, `libflutter.so`) |
| `smali/` | Decompiled Dalvik bytecode |
| `apktool.yml` | APKTool configuration metadata |

<a id="phase-11"></a>
## Phase 11 — Final Security Checklist

- [ ] Hardcoded API keys
- [ ] Hardcoded passwords
- [ ] Plain-text tokens
- [ ] Debuggable build left enabled
- [ ] Unnecessarily exported components
- [ ] SSL/certificate pinning present
- [ ] Backup enabled (`allowBackup`)
- [ ] Cleartext HTTP traffic allowed
- [ ] WebView debugging enabled
- [ ] Sensitive data in logs
- [ ] Local storage encrypted
